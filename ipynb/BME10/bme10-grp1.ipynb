```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import pardir, path
%matplotlib inline

plt.rcParams["font.family"] = ["Loma", 'Angsana New', 'TH SarabunPSK']
```

```python
def dp(dataFileName):
    '''
    This function is used to get full data path (dp) from data file name.
    '''
    return path.join(pardir, pardir, 'data', dataFileName)
df = pd.read_csv(dp('Pokemon.csv'), encoding='utf-8')
# import Data to array df
```

```python
df
# Show !
```

```python
to_plot_df = df.groupby(['Type 1']).mean()

to_plot_df
#Group data by Type 1 and Show!
```

```python
# We plot desire column by attack 
pt = to_plot_df['Attack'].plot(kind='bar', figsize=(18,8), fontsize=12)
plt.xlabel(pt.get_xlabel(), fontsize=22)
plt.ylabel('Average Atk of each type', fontsize=22)
plt.title('Average Attack of each type', fontsize=30)
```

```python
result =  to_plot_df.sort_values(by=['Attack'],ascending=False)
result
#Sort them by attack
```

```python
pt = result['Attack'].plot(kind='bar', figsize=(18,8), fontsize=12)
plt.xlabel(pt.get_xlabel(), fontsize=22)
plt.ylabel('Average Atk of each type', fontsize=22)
plt.title('Average Attack of each type', fontsize=30)
#then show data again!
```
