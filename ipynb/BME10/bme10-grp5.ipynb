import library และ ทำให้อ่านภาษาไทยได้

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import path, pardir
%matplotlib inline
plt.rcParams["font.family"] = ["Loma", 'Angsana New', 'TH SarabunPSK']
```

ดึงข้อมูลออกมาและแสดงผล

```python
def dp(dataFileName):
    '''
    This function is used to get full data path (dp) from data file name.
    '''
    return path.join(pardir, 'data', dataFileName)
df = pd.read_csv(dp('travel_amount_2558.csv'))
df
```

ใส่ index ให้หัวข้อตารางข้อมูล

```python
# look at list of columns name below and select what to plot
for i, item in enumerate(df.columns.unique().tolist()):
    print(i, item)
```

แสดงหัวข้อตารางเป็นlist

```python
df.columns.unique().tolist()
# df['ปริมาณการเดินทางรวม\nTotal (PCU-กิโลเมตร)\n(PCU-kilometer)']
# df['รถบรรทุกเล็ก 4 ล้อ\nLigth Truck']
# new_df = df.apply(pd.to_numeric, errors='ignore')
```

ตัดrow และ column ที่มีข้อมูลไม่ใช่ตัวเลข และตัดข้อมูลรวมและปริมาณการเดินทางรวมออก เพื่อไม่ให้ข้อมูลมันพุ่งกว่าข้อมูลอันอื่น

```python
ignore = [17, 50, 54]
new_df = pd.concat([df[:17], df[18:50], df[51:54], df[55:]]).drop([ 'จำนวนของช่องจราจร\nNumber of lane'], axis=1)
new_df =new_df.drop([ 'รวม\nTotal'], axis=1).drop([ 'ปริมาณการเดินทางรวม\nTotal (PCU-กิโลเมตร)\n(PCU-kilometer)'], axis=1)
new_df = new_df.apply(pd.to_numeric, errors='ignore')
```

Checkว่า ข้อมูลเป็นตัวเลขแล้วหรือยัง ถ้าเป็น dttypeจะเป็น int64 ถ้าเป็นตัวหนังสือ จะเป็น Object

```python
new_df['รถยนต์นั่งไม่เกิน 7 คน\nPassenger Car. less  than 7 persons']
```

แสดงผลรวมของข้อมูลในแต่ละcolumn ที่มีหมายเลขสำนักทางหลวงเดียวกัน

```python
to_plot_df = new_df.groupby(['หมายเลขสำนักทางหลวง (Buraeu No.)']).sum()

to_plot_df = to_plot_df.iloc[:-2]

to_plot_df
```

Plot กราฟแท่งจากตารางข้อมูลข้างบน

```python
pt = to_plot_df.plot(kind='bar', figsize=(18,8), fontsize=12)
plt.xlabel('หมายเลขสำนักทางหลวง', fontsize=22)
plt.ylabel('ปริมาณการเดินทางรวม ', fontsize=22)
plt.title('ปริมาณการเดินทางรวม Total (PCU-กิโลเมตร) (PCU-kilometer)', fontsize=30)
```
